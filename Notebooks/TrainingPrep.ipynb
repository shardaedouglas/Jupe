# Training Prep
## This file will act as the initial training ground for the ADDIS Model. It is experimental and is exploring the capabilities of Isolation Forests

```python
# Import Libraries

import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import OneHotEncoder
```

```python
# Define Variables 

ghcn_cleaned = ""
```

```python
# Import Dataset

ghcn_cleaned = pd.read_csv("../Datasets/GHCN_Data/Training_Data/ghcn_cleaned.csv")
# ghcn_cleaned = pd.read_csv("ghcn_cleaned.csv")

```

```python
# Hot encode categorical columns

categorical_columns = ghcn_cleaned.select_dtypes(include=['object']).columns.tolist()
categorical_columns
encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(ghcn_cleaned[categorical_columns])
hot_ghcn = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))
ghcn_encoded = pd.concat([ghcn_cleaned, hot_ghcn], axis=1)
ghcn_encoded = ghcn_encoded.drop(categorical_columns, axis=1)
ghcn_encoded
```

```python
# Instantiate the IsolationForest Model

model = IsolationForest(contamination='auto', random_state=42)
```

```python
# Train the Model

# Fit the model to the dataset

model.fit(ghcn_encoded)
```
